```python
import spacy
```

```python
def read_text_file(filename):
    file = open(filename, "r", encoding="utf-8") 
    return file.read()
```

```python
# upload NLP_syllabus_blog_post.txt to colab
from google.colab import files
uploaded = files.upload()
```

```python
text = read_text_file("NLP_syllabus_blog_post.txt")
```

```python
model = spacy.load('en_core_web_sm')
```

```python
doc = model(text)
```

```python
for noun_chunk in doc.noun_chunks:
    print(noun_chunk.text)
```

```python
s = "I am currently starting work on developing an undergraduate module in Natural Language Processing (level 6, 3rd year)"
```

```python
doc = model(s)
```

```python
for noun_chunk in doc.noun_chunks:
    print(noun_chunk.text)
```

```python
for noun_chunk in doc.noun_chunks:
    print(noun_chunk.text, "\t", noun_chunk.start, "\t", noun_chunk.end)
```

```python
for noun_chunk in doc.noun_chunks:
    print(noun_chunk.text, "\t", noun_chunk.start, "\t", noun_chunk.sent)
```

```python
for noun_chunk in doc.noun_chunks:
    print(noun_chunk.text, "\t", noun_chunk.start, "\t", noun_chunk.root.text)
```

```python
sent2 = "computational linguistics"
doc2 = model(sent2)
```

```python
for noun_chunk in doc.noun_chunks:
    print(noun_chunk.text, "\t", noun_chunk.similarity(doc2))
```
