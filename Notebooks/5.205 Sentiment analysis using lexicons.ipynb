## Sentiment analysis using lexicons

### 1. TextBlob

```python
import textblob
from textblob import TextBlob
```

```python
sentences = ["the book is good", "one of my professional favourites", "examples are not comprehensive", "images were not good", "book was not good"]
```

```python
def get_blob_sentiment(sentence):
    result = TextBlob(sentence).sentiment
    print(sentence, result.polarity)
    return result.polarity
```

```python
for sentence in sentences:
    sentiment = get_blob_sentiment(sentence)
```

### 2. nltk

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
sid = SentimentIntensityAnalyzer()
```

```python
def get_nltk_sentiment(sentence):
    ss = sid.polarity_scores(sentence)
    print(sentence, ss)
    return ss['compound']
```

```python
for sentence in sentences:
    sentiment = get_nltk_sentiment(sentence)
```

```python
a = get_nltk_sentiment(
    'book page complete junk could even finish page something got frustrated horrible threw floor ca even explain bad trust one unless english familiar english lingo hard understand'
)

print(a)
```
