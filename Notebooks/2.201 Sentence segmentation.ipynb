# Sentence segmentation
* using nltk http://www.nltk.org/
* an open source platform for working with text data
* provides access to many lexical resources and libraries for all sorts of NLP tasks

```python
# nltk is an open source platform for working with text data
# provides access to many lexical resources and libraries for all sorts of NLP tasks
import nltk
```

```python
# punkt is a pre-trained sentence tokenizer for English
nltk.download('punkt')
```

```python
text = '''Punctuation denotes the end of a sentence!
“But not always!”, said Fred... (who is a professor).
Some sentences could break over multiple lines
and begin with a lower case letter.
a1 could be a variable name.
'''
```

```python
text
```

```python
sents = nltk.sent_tokenize(text)
```

```python
sents[1]
```

```python
len(sents)
```

```python
for s in sents:
    print(s)
```

```python
print('\n----\n'.join(sents))
```
